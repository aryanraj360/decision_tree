```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
dataset.keys()
```

```python
df['target'] = dataset.target
```

```python
df.head()
```

```python
dataset.target_names
```

```python
df.info()
```

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
y
```

```python
## Train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)
```

```python
## Postpruning 
classifier = DecisionTreeClassifier(criterion='entropy', max_depth=2)
classifier.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled='True')
```

```python
## Prediction 
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test,y_pred)
```

### Decision Tree Prepruning And HyperParameter Tuning For Huge Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier=DecisionTreeClassifier()
```

```python
clf=GridSearchCV(classifier, param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
classifier=DecisionTreeClassifier(criterion='entropy', max_depth=4, max_features='sqrt', splitter='best')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
